```python
import pandas as pd
import functions as func
import yaml

try:
    with open("../config.yaml", "r") as file:
        config = yaml.safe_load(file)
except:
    print("The config.yaml file was not found in the main folder!")

df = pd.read_csv(config['data_raw']['file3'])

df
```

```python
#droping trash columns

df = df.drop(["Country Name","Indicator Name","Indicator Code","Unnamed: 68"],axis=1)
```

```python
#remaning col
df = df.rename(columns={'Country Code': 'country_code'})
```

```python
#drop years that wont be used

#cleaning 19**
for x in range(6,10):
    for y in range(1,10):
        drop_year = str(19)+str(x)+str(y)
        df.drop(drop_year, axis=1, inplace= True)

```

```python

#cleaning 20**
for x in range(0,2):
    for y in range(1,10):
        drop_year = str(20)+str(x)+str(y)
        df.drop(drop_year, axis=1, inplace= True)
```

```python
#missing 3 columns

df.drop(["2021", "2022", "2023"], axis=1, inplace= True)
```

```python
#drop missing values
df.dropna(inplace=True)
```

```python
#correcting data types for colums
df["1960"] = df["1960"].astype(int)
df["1970"] = df["1970"].astype(int)
df["1980"] = df["1980"].astype(int)
df["1990"] = df["1990"].astype(int)
df["2000"] = df["2000"].astype(int)
df["2010"] = df["2010"].astype(int)
df["2020"] = df["2020"].astype(int)
```

```python
# droping index column and set country_name as index
df = df.set_index("country_code")
```
