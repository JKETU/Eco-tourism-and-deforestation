```python
import pandas as pd
import functions as func
import yaml

try:
    with open("../config.yaml", "r") as file:
        config = yaml.safe_load(file)
except:
    print("The config.yaml file was not found in the main folder!")

df = pd.read_csv(config['data_raw']['file3'])

df
```

```python
#droping trash columns

df = df.drop(["Country Name","Indicator Name","Indicator Code","Unnamed: 68"],axis=1)
```

```python
#remaning col
df.rename(columns={'Country Code': 'country_code'})
```

```python
#drop years that wont be used

#cleaning 19**
for x in range(6,10):
    for y in range(1,10):
        drop_year = str(19)+str(x)+str(y)
        df.drop(drop_year, axis=1, inplace= True)

```

```python
df
```

```python

#cleaning 20**
for x in range(0,2):
    for y in range(1,10):
        drop_year = str(20)+str(x)+str(y)
        df.drop(drop_year, axis=1, inplace= True)
```

```python
#missing 3 columns

df.drop(["2021", "2022", "2023"], axis=1, inplace= True)
```

```python
#saving cleaned data frame
df.to_csv('../data/clean/tourism_data.csv')
```
